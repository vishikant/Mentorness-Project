```python
#Feature Selection
X_train=train_data.drop(columns=['Disease'])
y_train=train_data['Disease']

X_test=test_data.drop(columns=['Disease'])
y_test=test_data['Disease']
```

```python
# #Feature Engineering #Feature Scaling
# from sklearn.preprocessing import StandardScaler
# scaler=StandardScaler()
# scaled_data=scaler.fit_transform(X)
plt.figure(figsize=(12, 8))
sns.heatmap(X_train, cmap='viridis', cbar=True)
plt.title('Heatmap of Data')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Split data into train and test sets
model=LogisticRegression()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)
```

```python
test_data["ypred"]=y_pred
test_data
```

```python
random_forest=RandomForestClassifier()
random_forest.fit(X_train,y_train)
random_forest_predict=random_forest.predict(X_test)
test_data['RandomForestPred']=random_forest_predict
test_data
```

```python
def evalute_model(y_true,y_pred):
    recall=recall_score(y_true,y_pred)
    f1=f1_score(y_true,y_pred)
    accuracy=accuracy_score(y_true,y_pred)
    precision=precision_score(y_true,y_pred)
    return recall,f1,accuracy,precision
```

```python
from sklearn.metrics import accuracy_score

# Predict on the testing data
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
#it is a multiclass problem thats y i have used average="weighted" instead of binary
precision = precision_score(y_test, y_pred, average='weighted')  
recall = recall_score(y_test, y_pred, average='weighted') 
f1 = f1_score(y_test, y_pred, average='weighted')  

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
